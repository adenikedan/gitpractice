# Git and GitHub for Data Scientists 

Git is an Open Source Distributed Version Control System. In other words, Git is a Distributed Version Control System which uses a remote repository which is stored in a server (could be GitHub) and a local repository which is stored in the computer of each developer.

More Understanding: Git is a version control system that lets you manage and keep track of your source code history. GitHub is a cloud-based hosting service that lets you manage Git repositories. If you have open-source projects that use Git, then GitHub is designed to help you better manage them.

# Commit Code

Git help us track what files are being changed, added, or deleted in real time in our repository in a software development process.

```python
#Let's first understand the very basics of the Git workflow. 
#A typical project is comprised of multiple files and folders. 
#In Git, you have to think in three parts: local files, committed files, and the staging area.
```

```python
#Your local files are your work-in-progress, also known as working copy. 
```

```python
#Then you have committed files. 
#When you add committed code to your local repo, you can choose to “push” to an upstream location. That upstream location is the server. 
#During the push, you can also resolve conflicts, merge your team members' code, etc.
```

```python
#Finally, you have an area that sits in the middle of the committed area, 
#and the working area, which is the staging area. The staging area is your “proposal 
#to commit.” Think of it as: Okay I've been working on stuff in my local area, 
#and I'm ready to commit. And here is my proposal of what I want to commit.
```

# Adding a README FILE

What is a README File?
In simple words, we can describe a README file as a guide that gives users a detailed description of a project you have worked on.

It can also be described as documentation with guidelines on how to use a project. Usually it will have instructions on how to install and run the project.

It is essential for you as a developer to know how to document your project by writing a README 

Components of a Read Me File:
1. Project's Title
2. Project Description
3. How to Install and Run the Project
4. How to Use the Project
5. Include Credits for Team Work or Resources you Consulted
6. Add a License (check:https://choosealicense.com/)
7. How to Contribute to the Project (for open source projects)
8. Snapshots of the working prototype

Next, you need to set the upstream branch called “main” branch, then push your “main” into “origin”, I use the following command.

```python
git branch -M main
```

Git remote lets me manage tracked remote repositories. By saying Git remote add, I'm saying that I wish to add a remote tracked repository. The "origin" keyword indicates that here is where this project will be originally cloned from, in future, other people/developers can clone from there.

```python
#Now back to my local Git repo. I wish to push my local code into the server-side repo. 
#How does my local repo know where to push to? The answer is that I need to add 
#a remote origin, and here's how you do it:

git remote add origin https://github.com/rabonni/learngit.git
```

```python
git push -u origin main

#It will ask you for your GitHub Authentication Username and Code
#If, validated, your code will be pushed to the origin in the main branch.
```

Checking the Result/Output above: So when it says 3 objects, it's actually counting and pushing the following 3 objects: 1) The Git Commit Object 2) The Object that corresponds to the requirements.txt file. 3) The Working Tree Object placed in the jupyter home called Git Repository (.git directory).

# Modify Code

Most real-world repositories set a policy on the Git repo to prevent making changes to the “main” branch. The idea is that you create an issue. Then you discuss what you wish to do on that issue. You associate the issue with the files you're changing and you create a separate branch for your changes.

You create a branch and you make your changes there. And then you “merge” your changes into “main” via a pull request.

E.g. Also your team members can also create different branches/aspects of the project, work on them and push them to the server.

```python
git branch newchange
```

```python
#This command has now effectively given you a copy of “main”. 
#Don't worry, it isn't literally a full copy. Git is smart enough to abstract 
#the details only for changes. For you, it feels like a copy. 
#Before you can start working on this copy, you need to check out this branch, as follows:

 
git checkout newchange

#I could have also abbreviated the above two commands into one, 
#effectively saying “create a branch and check it out” like this:

 
git checkout -b newchange
```

```python
git branch #to verify which branch you are on.
```

```python
#Now that you've checked out the newchange branch, 
#let's make some changes. Modify the first file by appending some text.

 
echo "more changes" >> About Me.txt
#And create a new file.
```

```python
#Now let's stage this change, commit it locally, and then push it into the cloud.

git add .

git commit -m "My second commit"

#Look at you, issuing Git commands like a pro. I'm so proud. Now let's push it to the cloud.

 
git push -u origin newchange #set-upstream-to (or -u in short)
```

# Understanding the Version Control

Your output from the above code should look like: "Your branch is up to date with 'origin/newchange'."
This means after your "commit", your local "origin" on your machine compared itself to the remote counterpart "newchange" in your server (github).

I effectively now have multiple versions of my code base. Isn't this great? I can now revert back to a production version in “main” while switching to a dev version in “newchange”.

```python
#This brings up a question. How do I get my changes from “newchange” into “main”?
git merge newchange main
#You can push this and your remote repo will reflect these changes.
```

```python
#You can feel free to also delete your local “newchange” branch as follows:
git branch -d newchange
```
